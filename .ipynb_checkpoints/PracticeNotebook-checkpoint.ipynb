```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

plt.figure(figsize=(5,3), dpi=300)

# plt.plot(x, y, label='2x', color='red', linewidth=1, marker='.', markersize=5, markeredgecolor='blue', linestyle='--')
# Use shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r.--', label='2x')

## Line Number two
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2[:5], x2[:5]**2, 'b', label='X^2')
plt.plot(x2[4:], x2[4:]**2, 'b--', label='X^2')

plt.title("Our first graph!", fontdict={'fontname': 'Times New Roman', 'fontsize': 20})

plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# plt.xticks([0,1,2,3,4,5,6,7])
# plt.yticks([0,2,4,6,8,10,12,14])

plt.legend()

plt.show()
```
