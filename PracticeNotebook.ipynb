```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line Graphs

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]

# Dpi specifies density of pixel per inch. Should try to use 300 or more if you want to print higher resolution
plt.figure(figsize=(5,3), dpi=100) # 5,3 is the ratio of x by y

## Line Number One

# Keyword Argument Notation
# plt.plot(x, y, label='2x', color='red', linewidth=1, marker='.', markersize=5, markeredgecolor='blue', linestyle='--')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x, y, 'r.--', label='2x')

## Line Number two

# Select interval you want to plot points at
x2 = np.arange(0, 4.5, 0.5)

# Plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2, 'b', label='X^2')

# Plot remainder of graph
plt.plot(x2[4:], x2[4:]**2, 'b--')

# Annotate your figure with some data
plt.title("Line Graphs", fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.xlabel("X Axis", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 10})
plt.ylabel("Y Axis")

# Specify the values marked on the two axes
# plt.xticks([0,1,2,3,4,5,6,7])
# plt.yticks([0,2,4,6,8,10,12,14])

# Add a legend
plt.legend()

# Save the generated graph/figure to your local machine/directory
plt.savefig('mygraph.png', dpi=300)

# Show the graph below
plt.show()
```

## Bar charts

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['*', 'O', '/']
for i, bar in enumerate(bars):
  bar.set_hatch(patterns[i])

# You can also the hatch values individually to each bar as follows:
# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize=(6,4))

plt.show()
```

## Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')

fifa.head(5)
```

### Histograms

```python
bins = [i for i in range(30,101,10)] # start = 0, end = 101, step = 10

plt.hist(fifa.Overall, bins=bins, color='r')

plt.xticks(bins)

plt.title('Distribution of Player Skills in Fifa')
plt.xlabel('Player Ratings')
plt.ylabel('Number of Players')

plt.show()
```
